```python
import pandas as pd
import os

data_dir = r"C:\Users\darre\DataProjects\data"
csv_files = [f for f in os.listdir(data_dir) if f.endswith('.csv')]

for file in csv_files:
    file_path = os.path.join(data_dir, file)
    df = pd.read_csv(file_path)

    print(f"\n=== {file} ===")
    print(f"Shape: {df.shape} (rows, columns)")
    print("\nFirst 3 rows:")
    print(df.head(3))
    print("\nData Types and Missing Values:")
    print(df.dtypes)
    print("\nMissing Values per Column:")
    print(df.isnull().sum())
    print("\nDuplicates:", df.duplicated().sum())

    # Example schema check (adjust for key columns)
    if 'order_id' in df.columns:
        print(f"Sample order_ids (unique: {df['order_id'].nunique()}): {df['order_id'].head().tolist()}")

    # Basic stats for numeric columns
    numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
    if len(numeric_cols) > 0:
        print("\nBasic Stats (Numeric):")
        print(df[numeric_cols].describe())

# Overall summary
print(f"\nTotal files verified: {len(csv_files)}")
```
